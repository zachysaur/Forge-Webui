# Install Controlnet Models

```python
import os
import requests

# Define the URL of the model
model_url = "https://huggingface.co/thibaud/controlnet-openpose-sdxl-1.0/resolve/main/OpenPoseXL2.safetensors?download=true"

# Define the directory to save the model
save_directory = "/kaggle/working/stable-diffusion-webui-forge/models/ControlNet"

# Create the directory if it doesn't exist
os.makedirs(save_directory, exist_ok=True)

# Define the filename
filename = os.path.join(save_directory, "Openpose.XL2.safetensors")

# Download the model
response = requests.get(model_url)
if response.status_code == 200:
    with open(filename, 'wb') as f:
        f.write(response.content)
    print("Model downloaded successfully.")
else:
    print("Failed to download the model.")
```
